```python
import matplotlib.pyplot as plt
import logomaker as lm
import pandas as pd
import numpy as np
import random

%matplotlib inline
plt.ion()



#awk '!/^>/ { printf "%s", $0; n = "\n" } /^>/ { print n $0; n = "" } END { printf "%s", n } ' output_aln.fasta > new1.fa

def plotRegions(df, title, start, end):
    lm.Logo(df[start:end])
    plt.title(f'Ab Region: {title} (IMGT)')
    plt.ylabel('Information (Bits)')
    plt.xlabel('Position')
    return 0

with open('new1.fa') as f:
	lines = [line.rstrip('\n') for line in f]

# extract ww domain sequences
seqs = [seq.strip().upper() for seq in lines if ('#' not in seq) and ('>') not in seq]

# Take samples
gen_seq = seqs[-100:] # Last 100 are generated
seqs = seqs[:-100]
seqs = random.sample(seqs, 100)

print (len(seqs))


# Preview sequences
print('There are %d sequences, all of length %d'%(len(seqs), len(seqs[0])))
# print (seqs[:10])

# create counts matrix
ww_counts_df = lm.alignment_to_matrix(sequences=seqs, to_type='information', characters_to_ignore='.-X')

ww_counts_df_gen = lm.alignment_to_matrix(sequences=gen_seq, to_type='information', characters_to_ignore='.-X')

# preview counts dataframe
print (ww_counts_df.head())

# info_mat = lm.transform_matrix(ww_counts_df, 
#                                       from_type='counts', 
#                                       to_type='information')

# lm.Logo(ww_counts_df)
# lm.Logo(ww_counts_df_gen)
```

```python
# filter base on counts
num_seqs = ww_counts_df.sum(axis=1)
print(num_seqs.head())
pos_to_keep = num_seqs > 0.5

# filter base on counts GEN
num_seqs_gen = ww_counts_df_gen.sum(axis=1)
print(num_seqs_gen.head())
pos_to_keep_gen = num_seqs_gen > 0.5
```

```python
# TR
ww_counts_df = ww_counts_df[pos_to_keep]
ww_counts_df.reset_index(drop=True, inplace=True)
# lm.Logo(ww_counts_df)
plotRegions(ww_counts_df, 'Train - All Regions', 0, 128)

# Gen
ww_counts_df_gen = ww_counts_df_gen[pos_to_keep_gen]
ww_counts_df_gen.reset_index(drop=True, inplace=True)
# lm.Logo(ww_counts_df_gen)
plotRegions(ww_counts_df_gen, 'Generated - All Regions', 0, 128)
```
